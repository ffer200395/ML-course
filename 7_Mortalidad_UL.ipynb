# Clustering en sklearn

```python
# Paso 1: Escalar las variables de interés
X = df.iloc[:,3:]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
print(X_scaled.shape)
```

```python
# Paso 2: Entrenar un k-means para dos grupos
kmeans = KMeans(n_clusters=2, random_state=0).fit(X_scaled)
kmeans.labels_
```

```python
kmeans.inertia_
```

```python
# Paso 3: Bucle de 1 a 20 y guardar las incercias
ls_inertia = list()
for i in range(1,21):
  kmeans = KMeans(n_clusters=i, random_state=42).fit(X_scaled)
  ls_inertia.append(kmeans.inertia_)
ls_inertia
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,21),ls_inertia)
plt.show()
```

```python
# Paso 4 entrenamos con k=5 y almacenamos la columna
kmeans = KMeans(n_clusters=5, random_state=42).fit(X_scaled)
df['labels'] = kmeans.labels_
```

# Clustering en pySpark

```python
from pyspark.sql import SparkSession
from google.colab import drive
import pyspark.sql.functions as F
```

```python
spark = SparkSession.builder.appName("Python Spark Mortalidad UL").getOrCreate()
```

```python
# Path
path = '/content/gdrive/MyDrive/Colab Notebooks/Formación Python y ML/prácticas/'
# Mount drive
drive.mount('/content/gdrive')
```

```python
df_spark = spark.read.options(delimiter=",", header=True, inferSchema=True).csv(path+'df_mortalidad_country.csv')
df_spark.printSchema()
df_spark.show()
```

```python
df_spark.columns[3:]
```

```python
from pyspark.ml.feature import VectorAssembler
# Creamos un vector para cada fila
assemble=VectorAssembler(inputCols=df_spark.columns[2:], outputCol='features')
assembled_data=assemble.transform(df_spark)
assembled_data.show(2)
```

```python
from pyspark.ml.feature import StandardScaler
scaler_spark = StandardScaler(inputCol='features', outputCol='features_scaled')
df_scaled = scaler_spark.fit(assembled_data).transform(assembled_data)
```

```python
df_scaled.show()
```

```python
from pyspark.ml.clustering import KMeans
# Find a good K
# "Silhouette Score using ClusteringEvaluator() measures how close each point in one cluster is to points in the
# neighboring clusters thus helping in figuring out clusters that are compact and well-spaced out."
elbow_score=[]
for i in range(2,50):
    KMeans_algo=KMeans(featuresCol='features_scaled', k=i)
    KMeans_fit=KMeans_algo.fit(df_scaled)
    cost = KMeans_fit.summary.trainingCost
    elbow_score.append(cost)
    print("Elbow Distance:",cost, i)
```

```python
fig, ax = plt.subplots(1,1, figsize =(8,6))
ax.plot(range(2,50),elbow_score)
ax.set_xlabel("Number of Clusters")
ax.set_ylabel("Score")
```

```python
KMeans_algo=KMeans(featuresCol='features_scaled', k=10)
df_scaled_label = KMeans_algo.fit(df_scaled).transform(df_scaled)
```
